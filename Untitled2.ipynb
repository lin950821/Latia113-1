```python
def test3(x):
  result=0
  while result<10:
    result=result+x
    if result==5:
      return result
  return result
```

```python
b=test3(1)
print(b)
```

因為輪迴一次加1,到5時會被if函式返回結果

```python
c=test3(2)
print(c)
```

因為輪迴一次加2,所以可以重複執行到return=10

```python
def test4(value):
  return value*2-1
```

```python
test_values = [0.0, 0.0011, 0.25, 0.5,0,54, 0.75, 1.0]
scaled_values = [test4(value) for value in test_values]
print("Original values: ", test_values)
print("Scaled values: ", scaled_values)
```

```python
def test5(a):
  if a==0 or a==1:
    return 1

  else:
    return a+test5(a-1)
```

```python
print(test5(5))
```

結果:5+a(4)=15

a(4)=4+a(3)=10

a(3)=3+a(2)=6

a(2)=2+a(1)=3

a(1)=1
